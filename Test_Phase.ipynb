# Python Intro Demo (for Prog1 exam)

This notebook contains a single self-contained demonstration of basic Python concepts useful for a programming 1 exam. It shows variables, control flow, data structures, functions, modules (inlined), and exception handling, and also includes tiny tests/examples.

```python
# --- Utility functions (inlined for notebook) ---

def factorial(n: int) -> int:
    """Return n! for n >= 0. Raises ValueError for negative input."""
    if n < 0:
        raise ValueError("factorial() not defined for negative values")
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result


def is_prime(n: int) -> bool:
    """Return True if n is a prime number (simple trial division)."""
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True


def average(nums):
    """Return the average of a non-empty list/iterable of numbers."""
    nums = list(nums)
    if len(nums) == 0:
        raise ValueError("average() requires a non-empty sequence")
    return sum(nums) / len(nums)


def greet(name: str) -> str:
    return f"Hello, {name}!"


# --- Demo functions ---

def demo_variables_and_types():
    print("--- Variables and Types ---")
    x = 10                    # integer
    y = 3.14                  # float
    name = "Alice"          # string
    is_student = True         # boolean
    print(f"x={x} (type={type(x).__name__}), y={y} (type={type(y).__name__})")
    print(f"name={name}, is_student={is_student}")


def demo_control_flow():
    print("--- Control Flow ---")
    for i in range(1, 6):
        if i % 2 == 0:
            kind = "even"
        else:
            kind = "odd"
        print(f"{i} is {kind}")


def demo_data_structures():
    print("--- Data Structures ---")
    numbers = [1, 2, 3, 4, 5]
    squares = [n * n for n in numbers]   # list comprehension
    info = {"name": "Bob", "age": 22}  # dict
    unique = set([1, 2, 2, 3])
    print("numbers:", numbers)
    print("squares:", squares)
    print("info:", info)
    print("unique set:", unique)


def demo_functions_and_modules():
    print("--- Functions & Modules ---")
    n = 7
    print(f"factorial({n}) =", factorial(n))
    print(f"is_prime({n}) =", is_prime(n))
    print("greeting:", greet("Exam Student"))


def demo_exception_handling():
    print("--- Exceptions ---")
    try:
        print("Average of empty list:", average([]))
    except ValueError as e:
        print("Caught an error:", e)


# --- Main demo runner ---

def main():
    print("Python Intro Demo — Prog1 exam helper\n")
    demo_variables_and_types()
    print()
    demo_control_flow()
    print()
    demo_data_structures()
    print()
    demo_functions_and_modules()
    print()
    demo_exception_handling()


# If run in notebook, call main()
main()


# --- tiny checks / examples students can run interactively ---
print('\nQuick checks:')
print('factorial(5) ->', factorial(5))
print('is_prime(13) ->', is_prime(13))
print('average([1,2,3]) ->', average([1, 2, 3]))
print(greet('Student'))
```
